```python
from pathlib import Path
from ultralytics import YOLO
from utils.download import download_file
from PIL import  Image
import cv2
from matplotlib import pyplot as plt

# On récupère le modèle Yolo
yolo_model = YOLO('yolov8n.pt')
```

```python

# Chargement de l'image
img_path = download_file("https://upload.wikimedia.org/wikipedia/commons/6/63/Horse-racing-4.jpg")

image = Image.open(str(img_path))
```

```python
# Prédiction par Yolo pour l'image
result = yolo_model.predict(image)

# Génération d'une image résultat incluant les boîtes
result_image = result[0].plot()
# Sauvegarde de l'image résultat dans un fichier
out_dir = Path("out")
out_dir.mkdir(parents=True,exist_ok=True)
cv2.imwrite(str(out_dir / img_path.name),result_image)

# Affichage de l'image résultat
fig = plt.figure(layout="tight")
result_image = cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB)
fig.gca().imshow(result_image)
fig.gca().axis('off')
```
